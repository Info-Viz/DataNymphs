```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
```

```python
#richiamo l'endpoint di wikidata 
wikidata_endpoint = "https://query.wikidata.org/sparql"
sparql_wd = SPARQLWrapper(wikidata_endpoint)
```

```python
#creiamo una funzione per eseguire le query e restituire i risultati in un dataframe

def execute_wikidata_query(query, category):
    sparql_wd.setQuery(query)
    sparql_wd.setReturnFormat(JSON)
    results = sparql_wd.query().convert()
    
    #estraggo i risultati e li metto in un dataframe
    data = []
    for r in results["results"]["bindings"]: #per ogni risultato estraggo i valori dei campi
        entity = r["personaggio"]["value"] if "personaggio" in r else r["evento"]["value"] #se c'è il campo personaggio prendo quello, altrimenti prendo evento

        label = r["persLabel"]["value"] if "persLabel" in r else r["eventLabel"]["value"] #se c'è il campo persLabel prendo quello, altrimenti prendo eventLabel

        aliases = r["aliases"]["value"] if "aliases" in r else None #se c'è il campo aliases prendo quello, altrimenti metto None

        description = r["descrizione"]["value"] if "descrizione" in r else None #se c'è il campo descrizione prendo quello, altrimenti metto None
    
        data.append({
            "entity": entity,
            "label": label,
            "aliases": aliases,
            "description": description,
            "category": category
        })

    df = pd.DataFrame(
        data, 
        columns=["entity", "label", "aliases", "description", "category"]
    )

    return df
```

```python
# QUERY PER GREEK DEITIES (743)
query_greek_deities = """
SELECT ?personaggio ?persLabel
       (GROUP_CONCAT(DISTINCT ?aka; separator=", ") AS ?aliases)
       ?descrizione

WHERE {

  VALUES ?classiAmmesse {
    wd:Q22989102
  }

  ?personaggio wdt:P31/wdt:P279* ?classiAmmesse .

  ?personaggio rdfs:label ?persLabel .
  FILTER(LANG(?persLabel) = "it")

  OPTIONAL {
    ?personaggio skos:altLabel ?aka .
    FILTER(LANG(?aka) = "it")
  }

  OPTIONAL {
    ?personaggio schema:description ?descrizione .
    FILTER(LANG(?descrizione) = "it")
  }

}

GROUP BY ?personaggio ?persLabel ?descrizione
ORDER BY ?persLabel
"""

#DF GREEK DEITIES
df_greek_deities = execute_wikidata_query(query_greek_deities, "greek_deity") 

print(len(df_greek_deities))
```

```python
#elimino i duplicati che non hanno la descrizione quando altri invece ce l'hanno

df_e = df_uniti.drop_duplicates(subset="entity").copy()

df_e["has_desc"] = df_e["description"].notna() & (df_e["description"].str.strip() != "")

def resolve_label(group):
    # CASO 1: almeno una descrizione → tengo solo quelle
    if group["has_desc"].any():
        return group[group["has_desc"]]

    # CASO 2: nessuna descrizione → tengo una sola riga
    return group.head(1)

df_finalissimo = (
    df_e.groupby("label", group_keys=False)
      .apply(resolve_label)
      .drop(columns=["has_desc"])
      .reset_index(drop=True)
)
```

```python
#estraggo tutte le label  
labels = (
    df_finalissimo["label"]
    .dropna() #tolgo i valori nulli
    .str.lower() #metto tutto in minuscolo
    .str.strip() #tolgo gli spazi bianchi all'inizio e alla fine
    .tolist() #converto in lista

)

#estraggo gli aliases separando quelli multipli
aliases = []
for alias_str in df_finalissimo["aliases"].dropna(): #per ogni stringa di alias che non è nulla
    alias_list = [alias.strip().lower() for alias in alias_str.split(",")] #separo gli alias con la virgola, tolgo gli spazi e metto tutto in minuscolo
    aliases.extend(alias_list) #aggiungo alla lista totale degli alias

#unisco tutto 
myth_terms = list(set(labels + aliases)) #unisco le label e gli alias, tolgo i duplicati con set e riconverto in lista
len(myth_terms)
```

```python
#ENDPOINT SPARLQ DI ARCO 
arco_endpoint = "https://dati.cultura.gov.it/sparql"

sparql_arco = SPARQLWrapper(arco_endpoint)
sparql_arco.setReturnFormat(JSON)
```

```python
# NUMERO TOTALE DEI BENI STORICO ARTISTICI - senza duplicati 
query1 = """
PREFIX arco: <https://w3id.org/arco/ontology/arco/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#> 

SELECT (COUNT(DISTINCT ?item) AS ?totalItems)
WHERE {
  ?item rdf:type <https://w3id.org/arco/ontology/arco/HistoricOrArtisticProperty> .
}
"""


sparql_arco.setQuery(query1)
results = sparql_arco.query().convert()

for result in results["results"]["bindings"]:
    print(f"Total Historic or Artistic Properties: {result['totalItems']['value']}")
```

```python
#query pr estrarre i dati dei beni storico artistici i batch da 20000  

query2 = f"""
PREFIX arco: <https://w3id.org/arco/ontology/arco/>
PREFIX dc: <http://purl.org/dc/elements/1.1/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX a-cd: <https://w3id.org/arco/ontology/context-description/>

SELECT DISTINCT ?item ?label ?title ?date ?description ?historical 
WHERE {{
  ?item rdf:type <https://w3id.org/arco/ontology/arco/HistoricOrArtisticProperty> .

  OPTIONAL {{ ?item rdfs:label ?label }}
  OPTIONAL {{ ?item dc:title ?title }}
  OPTIONAL {{ ?item dc:date ?date }}
  OPTIONAL {{ ?item dc:description ?description }}
  OPTIONAL {{ ?item a-cd:historicalInformation ?historical }}
}}

LIMIT 10000
OFFSET 0

"""

sparql_arco.setQuery(query2)
results = sparql_arco.query().convert()
```

```python
rows = []

for r in results["results"]["bindings"]:

    rows.append({
        "item": r["item"]["value"] if "item" in r else None,
        "label": r["label"]["value"] if "label" in r else None,
        "title": r["title"]["value"] if "title" in r else None,
        "date": r["date"]["value"] if "date" in r else None,
        "description": r["description"]["value"] if "description" in r else None,
        #"historical": r["historical"]["value"] if "historical" in r else None
    })

df_arco = pd.DataFrame(rows)
len(df_arco)
#df_arco.head(10)
```

```python
# nel dataframe uniamo i campi di title e label in un unico campo "name" per facilitare la ricerca
df_arco["search_text"] = (
    df_arco["label"].fillna("") + " " + df_arco["title"].fillna("")
).str.lower()

df_arco[["label", "title", "search_text"]].head(1)
```

```python
# Creiamo search_text solo da label e title
df_arco["search_text"] = (
    df_arco["label"].fillna("") + " " +
    df_arco["title"].fillna("")
).str.lower()

import re

# Funzione per trovare tutti i termini che matchano ESATTAMENTE
def find_matched_terms(text, terms_list):
    """Cerca tutti i termini che compaiono nel testo come parole esatte"""
    matched = []
    for term in terms_list:
        # Usa word boundaries \b per cercare la parola esatta
        if re.search(r'\b' + re.escape(term) + r'\b', text):
            matched.append(term)
    return ", ".join(matched) if matched else None

# Applica il matching
df_arco["matched_term"] = df_arco["search_text"].apply(
    lambda text: find_matched_terms(text, myth_terms)
)

# Filtra solo le righe con almeno un match
df_matches = df_arco[df_arco["matched_term"].notna()].copy()
print(f"Righe con almeno 1 match: {len(df_matches)}")
```
