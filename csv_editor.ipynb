## 1️⃣ Load and Display the Data

````python
import pandas as pd
import os

# Carica il CSV
csv_path = 'df_matches_sorted.csv'
df = pd.read_csv(csv_path)

print(f"✅ CSV caricato con successo!")
print(f"Dimensioni: {df.shape[0]} righe, {df.shape[1]} colonne\n")

# Visualizza le prime righe
print("Primissime righe:")
df.head(10)
````

## 2️⃣ Explore Data Structure

````python
print("Nomi delle colonne:")
print(df.columns.tolist())
print("\n" + "="*60)

print("\nTipi di dati:")
print(df.dtypes)
print("\n" + "="*60)

print("\nStatistiche:")
print(df.info())
print("\n" + "="*60)

print("\nValori mancanti (NaN):")
print(df.isnull().sum())
print("\n" + "="*60)

print("\nRighe duplicate:")
print(f"Totale duplicati: {df.duplicated().sum()}")
````

## 3️⃣ Filter and Search Data

````python
# Esempio 1: Cercare una stringa in una colonna
def search_in_column(column_name, search_term):
    """Cerca una stringa specifica in una colonna"""
    result = df[df[column_name].astype(str).str.contains(search_term, case=False, na=False)]
    return result

# Esempio 2: Filtrare per valore esatto
def filter_by_value(column_name, value):
    """Filtra le righe dove una colonna ha un valore specifico"""
    result = df[df[column_name] == value]
    return result

# Esempio 3: Filtrare per valori multipli
def filter_by_values(column_name, values_list):
    """Filtra le righe dove una colonna contiene uno dei valori"""
    result = df[df[column_name].isin(values_list)]
    return result

# Esempio 4: Filtrare per condizione numerica
def filter_by_condition(column_name, operator, value):
    """Filtra per condizioni numeriche (>, <, >=, <=, ==)"""
    if operator == '>':
        result = df[df[column_name] > value]
    elif operator == '<':
        result = df[df[column_name] < value]
    elif operator == '>=':
        result = df[df[column_name] >= value]
    elif operator == '<=':
        result = df[df[column_name] <= value]
    elif operator == '==':
        result = df[df[column_name] == value]
    return result

print("Funzioni di filtro disponibili:")
print("- search_in_column('colonna', 'testo')")
print("- filter_by_value('colonna', valore)")
print("- filter_by_values('colonna', [val1, val2, ...])")
print("- filter_by_condition('colonna', '>', valore)")
print("\nEsempio: search_in_column('item', 'apollo')")
print("\nProva le funzioni qui sotto!")
````

## 4️⃣ Data Manipulation and Cleaning

````python
# Esempio 1: Rimuovere spazi bianchi
def clean_whitespace(column_name):
    """Rimuove spazi bianchi all'inizio e alla fine"""
    df[column_name] = df[column_name].str.strip()
    print(f"✅ Spazi rimossi da '{column_name}'")

# Esempio 2: Convertire a maiuscole/minuscole
def to_lowercase(column_name):
    df[column_name] = df[column_name].str.lower()
    print(f"✅ '{column_name}' convertita a minuscoli")

def to_uppercase(column_name):
    df[column_name] = df[column_name].str.upper()
    print(f"✅ '{column_name}' convertita a maiuscoli")

# Esempio 3: Rimuovere valori NaN
def remove_missing_values(column_name=None):
    """Rimuove righe con valori NaN"""
    if column_name:
        rows_before = len(df)
        df.dropna(subset=[column_name], inplace=True)
        rows_after = len(df)
        print(f"✅ Rimossi {rows_before - rows_after} righe con NaN in '{column_name}'")
    else:
        rows_before = len(df)
        df.dropna(inplace=True)
        rows_after = len(df)
        print(f"✅ Rimossi {rows_before - rows_after} righe con qualsiasi NaN")

# Esempio 4: Rimuovere duplicati
def remove_duplicates(subset=None):
    """Rimuove righe duplicate"""
    rows_before = len(df)
    df.drop_duplicates(subset=subset, inplace=True)
    rows_after = len(df)
    print(f"✅ Rimossi {rows_before - rows_after} righe duplicate")

# Esempio 5: Rinominare colonne
def rename_column(old_name, new_name):
    df.rename(columns={old_name: new_name}, inplace=True)
    print(f"✅ Colonna '{old_name}' rinominata in '{new_name}'")

# Esempio 6: Sostituire valori
def replace_values(column_name, old_value, new_value):
    df[column_name] = df[column_name].replace(old_value, new_value)
    print(f"✅ '{old_value}' sostituito con '{new_value}' in '{column_name}'")

print("Funzioni di pulizia disponibili:")
print("- clean_whitespace('colonna')")
print("- to_lowercase('colonna')")
print("- to_uppercase('colonna')")
print("- remove_missing_values('colonna') o remove_missing_values()")
print("- remove_duplicates(subset=['colonna'])")
print("- rename_column('vecchio_nome', 'nuovo_nome')")
print("- replace_values('colonna', 'vecchio', 'nuovo')")
````

## 5️⃣ Remove Rows

````python
# Esempio 1: Rimuovere per indice
def remove_row_by_index(index):
    """Rimuove una riga specifica per indice"""
    if index in df.index:
        df.drop(index, inplace=True)
        df.reset_index(drop=True, inplace=True)
        print(f"✅ Riga {index} rimossa!")
    else:
        print(f"❌ Indice {index} non trovato")

# Esempio 2: Rimuovere per indici multipli
def remove_rows_by_indices(indices):
    """Rimuove più righe specifiche"""
    df.drop(indices, inplace=True, errors='ignore')
    df.reset_index(drop=True, inplace=True)
    print(f"✅ {len(indices)} righe rimosse!")

# Esempio 3: Rimuovere dove una colonna corrisponde a un valore
def remove_rows_by_value(column_name, value):
    """Rimuove tutte le righe dove una colonna ha un valore specifico"""
    rows_before = len(df)
    df.drop(df[df[column_name] == value].index, inplace=True)
    df.reset_index(drop=True, inplace=True)
    rows_after = len(df)
    print(f"✅ {rows_before - rows_after} righe rimosse dove {column_name} == {value}")

# Esempio 4: Rimuovere dove una colonna contiene un testo
def remove_rows_by_text(column_name, text):
    """Rimuove tutte le righe dove una colonna contiene un testo"""
    rows_before = len(df)
    mask = df[column_name].astype(str).str.contains(text, case=False, na=False)
    df.drop(df[mask].index, inplace=True)
    df.reset_index(drop=True, inplace=True)
    rows_after = len(df)
    print(f"✅ {rows_before - rows_after} righe rimosse dove {column_name} contiene '{text}'")

# Esempio 5: Visualizzare righe prima di rimuovere
def preview_rows_to_delete(column_name, value):
    """Mostra le righe che verranno eliminate"""
    mask = df[column_name] == value
    result = df[mask]
    print(f"Queste {len(result)} righe verranno eliminate:\n")
    return result

print("Funzioni di rimozione disponibili:")
print("- remove_row_by_index(5)  # Rimuove la riga con indice 5")
print("- remove_rows_by_indices([1, 3, 5])  # Rimuove più righe")
print("- remove_rows_by_value('colonna', valore)")
print("- remove_rows_by_text('colonna', 'testo')")
print("- preview_rows_to_delete('colonna', valore)  # Anteprima prima di eliminare")
print("\n⚠️ CONSIGLIO: Usa preview_rows_to_delete() prima di eliminate!")
````

## 6️⃣ Save Modified Data

````python
# Opzione 1: Sovrascrivere il file originale
def save_to_original_file():
    """Salva i dati modificati sovrascrivendo il file originale"""
    df.to_csv(csv_path, index=False)
    print(f"✅ Dati salvati in '{csv_path}'")
    print(f"   Dimensioni finali: {df.shape[0]} righe, {df.shape[1]} colonne")

# Opzione 2: Salvare come nuovo file
def save_to_new_file(filename):
    """Salva i dati modificati in un nuovo file"""
    df.to_csv(filename, index=False)
    print(f"✅ Dati salvati in '{filename}'")
    print(f"   Dimensioni finali: {df.shape[0]} righe, {df.shape[1]} colonne")

# Opzione 3: Salvare con timestamp
from datetime import datetime

def save_with_timestamp():
    """Salva i dati con timestamp nel nome"""
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"df_matches_sorted_modified_{timestamp}.csv"
    df.to_csv(filename, index=False)
    print(f"✅ Dati salvati in '{filename}'")

# Opzione 4: Visualizzare stato attuale prima di salvare
def show_changes_summary():
    """Mostra un riassunto delle modifiche"""
    print("📊 RIASSUNTO MODIFICHE:")
    print(f"   Righe totali: {df.shape[0]}")
    print(f"   Colonne totali: {df.shape[1]}")
    print(f"   Colonne: {', '.join(df.columns.tolist())}")
    print(f"\n   ⚠️ Prima di salvare, verifica che tutto sia corretto!")

print("Funzioni di salvataggio disponibili:")
print("- save_to_original_file()  # Sovrascrive il file originale")
print("- save_to_new_file('nuovo_file.csv')")
print("- save_with_timestamp()  # Salva con data e ora nel nome")
print("- show_changes_summary()  # Visualizza riepilogo modifiche")
print("\n💡 CONSIGLI:")
print("   1. Usa show_changes_summary() prima di salvare")
print("   2. Inizialmente salva come nuovo file per non perdere l'originale")
print("   3. Dopo aver verificato, puoi sovrascrivere l'originale se necessario")
````
